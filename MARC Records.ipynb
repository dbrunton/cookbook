## Reading MARC records in Python

This Python library was written by someone who used to work at LC, Ed Summers!

```python
from pymarc import MARCReader
```

I also like to format the output of my Python functions, here's what I use for that:

```python
from IPython.display import Markdown, display
```

Note: "datasets/python.mrc" is a file of Marc records I downloaded from catalog.loc.gov and uploaded to GitHub after searching for the word "python".

```python
with open('datasets/python.mrc', 'rb') as fh:
    reader = MARCReader(fh)
    for record in reader:
        display(Markdown("__" + record.title() + "__ " + record.pubyear())) ## OR record['245']['a']
        for subject in record.subjects():
            print(subject)
        print("\n")
```
